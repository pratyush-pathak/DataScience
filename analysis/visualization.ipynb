```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_depth_mm'].plot()
```

```python
penguins['bill_depth_mm'].plot(linestyle='', marker='s', markersize=7, alpha=1, figsize=(20,10), markerfacecolor ='yellow', markeredgecolor = 'red')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar', color='green', figsize=(15,7))
```

```python
penguins.plot(kind='scatter', x='bill_length_mm', y='bill_depth_mm')
```

subplots


```python
fig, axs = plt.subplots(1, 2, figsize=(10,5), constrained_layout=True)
penguins.plot(kind='scatter', x='bill_length_mm', y='bill_depth_mm', ax = axs[0],title='penguin bill data')
penguins['bill_depth_mm'].plot(linestyle='', marker='s', markersize=7, alpha=1, figsize=(20,10), markerfacecolor ='yellow', markeredgecolor = 'red', ax=axs[1], title='penguin flipper')
plt.show()
```
