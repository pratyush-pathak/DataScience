```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_depth_mm'].plot()
```

```python
penguins['bill_depth_mm'].plot(linestyle='', marker='s', markersize=7, alpha=1, figsize=(20,10), markerfacecolor ='yellow', markeredgecolor = 'red')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar', color='green', figsize=(15,7))
```

```python
penguins.plot(kind='scatter', x='bill_length_mm', y='bill_depth_mm')
```

subplots


```python
fig, axs = plt.subplots(1, 2, figsize=(10,5), constrained_layout=True)
penguins.plot(kind='scatter', x='bill_length_mm', y='bill_depth_mm', ax = axs[0],title='penguin bill data')
penguins['bill_depth_mm'].plot(linestyle='', marker='s', markersize=7, alpha=1, figsize=(20,10), markerfacecolor ='yellow', markeredgecolor = 'red', ax=axs[1], title='penguin flipper')
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.countplot(data=tips, x='sex')
```

```python
sns.countplot(data=tips, x='sex', hue='day')
```

```python
sns.displot(data=tips, x='total_bill', binwidth=5, row='sex', col='smoker')  #distribution plot (histogram)
```

```python
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

```python
sns.set_theme(style="ticks")

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("anscombe")

# Show the results of a linear regression within each dataset
sns.lmplot(
    data=df, x="x", y="y", col="dataset", hue="dataset",
    col_wrap=2, palette="muted", ci=None,
    height=4, scatter_kws={"s": 50, "alpha": 1}
)
```
